**Description:**

Bayesian inference is a statistical method that updates the belief about a parameter based on observed data. It is built on **Bayes' theorem**, which states:

$$
p(\theta | D) = \frac{p(D | \theta) p(\theta)}{p(D)}
$$

where:
- $p(\theta | D)$ is the **posterior** distribution (our updated belief about $\theta$ after seeing data $D$),
- $p(D | \theta)$ is the **likelihood** (the probability of observing $D$ given $\theta$),
- $p(\theta)$ is the **prior** distribution (our belief about $\theta$ before seeing data),

Bayesian inference is widely used in machine learning, probabilistic modeling, and decision-making under uncertainty.


### **Frequentist vs. Bayesian**
| Aspect            | Frequentist Inference            | Bayesian Inference |
|------------------|---------------------------------|--------------------|
| **Parameters**   | Fixed but unknown               | Random variables  |
| **Probability**  | Long-run frequency of events    | Degree of belief  |
| **Data Role**    | Only current data is used       | Prior + data combined  |
| **Key Tool**     | Likelihood | Bayes' theorem |

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

seed = 1
np.random.seed(seed)
```

#### Exercice 11.2

```python
# Exercice 11.2
print(f"Exercice 11.2")

true_mu = 5
std = 1
num_samples = 100

data = np.random.normal(true_mu, std, num_samples)

print("*****************************************************************************************")
print("Question b:")
def posterior_density(mu, data):
    likelihood = np.exp(-0.5*np.sum((data - mu)**2))
    return likelihood

mu_s = np.arange(4.5, 5.75, 0.01)
posterior_values = np.array([posterior_density(mu, data) for mu in mu_s])
posterior_values /= np.sum(posterior_values)
print(f"{posterior_values.shape = }")

posterior_mode = mu_s[np.argmax(posterior_values)]
print(f"Posterior mode: {posterior_mode:.3}")

sns.set(style="whitegrid")
plt.figure(figsize=(8, 5))
sns.lineplot(x=mu_s, y=posterior_values, label='Posterior Density', color='blue')
plt.xlabel('Mu')
plt.ylabel('Density')
plt.title('Posterior Density for Mu')
plt.legend()
plt.show()

print("*****************************************************************************************")
print("Question c:")
posterior_samples = np.random.choice(mu_s, size=1000, p=posterior_values)
plt.figure(figsize=(8, 6))
sns.histplot(posterior_samples, bins=10, label='Simulated Samples')
plt.legend()
plt.show()

plt.figure(figsize=(8, 6))
sns.histplot(posterior_samples, bins=10, label='Simulated Samples')
sns.lineplot(x=mu_s, y=posterior_values*6.5e3,
             color='red', label='Posterior Density')
plt.xlabel('Mu')
plt.ylabel('Frequency')
plt.title('Histogram of Simulated Posterior Samples vs Posterior Density')
plt.legend()
plt.show()
print(f"The histogram should match the posterior density curve.")

print("*****************************************************************************************")
print("Question d:")
# Analytical approach
theta_s = np.exp(mu_s)
posterior_theta_analytical = posterior_values/theta_s
posterior_theta_analytical /= np.sum(posterior_theta_analytical)

# Simulation approach
posterior_samples_mu = np.random.choice(mu_s, size=1000, p=posterior_values)
posterior_samples_theta = np.exp(posterior_samples_mu)

sns.set(style="whitegrid")
plt.figure(figsize=(8, 6))
sns.lineplot(x=theta_s, y=posterior_theta_analytical*2e3, label='Analytical Posterior Density', color='red')
sns.histplot(posterior_samples_theta, bins=30, kde=False, color='skyblue', label='Simulated Posterior Samples')
plt.xlabel(r'$\theta$')
plt.ylabel('Density')
plt.title('Posterior Density for $\Theta$ (Analytical and Simulation)')
plt.legend()
plt.show()

print("*****************************************************************************************")
print("Question e:")
posterior_values = np.array([posterior_density(mu, data) for mu in mu_s])
posterior_values /= np.sum(posterior_values)
posterior_samples_mu = np.random.choice(mu_s, size=1000, p=posterior_values)
lower_bound, upper_bound = np.percentile(posterior_samples_mu, [2.5, 97.5])
print(f"95% Posterior Interval for : [{lower_bound:.2f}, {upper_bound:.2f}]")

print("*****************************************************************************************")
print("Question f:")
posterior_samples_theta = np.exp(posterior_samples_mu)
lower_bound, upper_bound = np.percentile(posterior_samples_theta, [2.5, 97.5])
print(f"95% Confidence Interval for : [{lower_bound:.2f}, {upper_bound:.2f}]")
```
